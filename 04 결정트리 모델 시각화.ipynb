```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore') #경고 무시
```

```python
# 데이터 로딩,분리
iris_data=load_iris()
X_train,X_test,y_train,y_test=\
train_test_split(
    iris_data.data
    ,iris_data.target
    ,test_size=0.2
    ,random_state=11
)
```

```python
# 모델 생성
dt_clf = DecisionTreeClassifier(random_state=156)

#학습데이터(답)으로 모델을 생성
dt_clf.fit(
    X_train
    ,y_train
)
```

```python
# 시각화에 필요한 dot 파일을 생성
from sklearn.tree import export_graphviz
export_graphviz(
      dt_clf #학습한 모델
    , out_file='tree.dot' #저장할 dot 파일 이름
    , class_names=iris_data.target_names
    , feature_names=iris_data.feature_names
    , impurity=True
    , filled=True
)
```

```python
# 그래프 시각화
import graphviz

# f=open('tree.dot','r')
# #처리 코드
# f.close()
with open('tree.dot') as f:
    dot_graph=f.read() #전체를 읽는다
    
graphviz.Source(dot_graph)
```

```python
# HAR 데이터셋을 가져오는 function
def get_human_dataset():
    # 피처명
    feature_name_df=\
    pd.read_csv(
        'D:/shinj/Data/[5]UCI-Human-Activity-Recognition/HAR/features.txt'
        ,sep='\s+'
        ,header=None
        ,names=['colum_index','colum_name']
    )
    fn = feature_name_df.iloc[:,1].values.tolist()
    
    X_train=pd.read_csv(
        'D:/shinj/Data/[5]UCI-Human-Activity-Recognition/HAR/train/X_train.txt'
        ,sep='\s+'
        ,header=None
    )
    
    X_test=pd.read_csv(
        'D:/shinj/Data/[5]UCI-Human-Activity-Recognition/HAR/test/X_test.txt'
        ,sep='\s+'
        ,header=None
    )
    
    y_train=pd.read_csv(
        'D:/shinj/Data/[5]UCI-Human-Activity-Recognition/HAR/train/y_train.txt'
        ,sep='\s+'
        ,names=['action']
    )
    
    y_test=pd.read_csv(
        'D:/shinj/Data/[5]UCI-Human-Activity-Recognition/HAR/test/y_test.txt'
        ,sep='\s+'
        ,names=['action']
    )
    
    return X_train,X_test,y_train,y_test
```

```python
X_train,X_test,y_train,y_test = get_human_dataset()
```

```python
#모델생성, 학습, 평가
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score #평가: 정확도
```

```python
#Model
dt_clf = DecisionTreeClassifier(random_state=156)
#fit
dt_clf.fit(X_train, y_train)
#pred
pred = dt_clf.predict(X_test)
#Acc
accu = accuracy_score(y_test, pred)
print('결정트리 예측 정확도 : {0:.4f}'.format(accu))
```

```python
#파라미터 디폴트 처리 -> 파라미터 확인
dt_clf.get_params()
```

```python
#'max_depth': None,-> 지정한게 없음
#교차검증, 베스트 파라미터 추출 : GridSearchCV
from sklearn.model_selection import GridSearchCV

params = {
    'max_depth' : [4,6,8,10,12,16,20,24]
}

#GridsearchCV객체생성
grid_cv = GridSearchCV(
    #1.적용할 모델 객체
    dt_clf
    #2. 적용할 파라미터
    ,param_grid=params
    #3. 적용할 평가지표
    ,scoring = 'accuracy'
    #4. 교차검증 갯수(비복원=중복추출x)
    ,cv=5
    ,verbose=1 #출력결과 형태지정, 0일때 안보임
)
#학습시작
grid_cv.fit(X_train, y_train)
```

```python
#정확도 출력 
grid_cv.best_score_
```

```python
#베스트 파라미터 출력
grid_cv.best_params_
```

```python
#베스트한 점수, 파라미터 값을 한번에 출력
cv_results_df = pd.DataFrame(grid_cv.cv_results_)
cv_results_df.columns
```

```python
cv_results_df[[ 'param_max_depth','mean_test_score']]
```

```python
#베스트 모델:grid_cv.best_estimator_ -> max_depth = 16
#파라미터 반복처리 정확도 출력
max_depths = [4,6,8,10,12,16,20,24]

for depth in max_depths:
    dt_clf = DecisionTreeClassifier(
        max_depth=depth, random_state=156
    )
    dt_clf.fit(X_train, y_train)
    pred = dt_clf.predict(X_test)
    accu = accuracy_score(y_test, pred)
    print('max_depth = {0} 정확도 : {1:4f}'.format(depth,accu))
```

```python
#parameter : max_depth, min_samples_split
params = {
     'max_depth' : [8,10,12,16,20] #8이상의 파라미터, 20과 24의 정확도는 같으므로 20한개만
    ,'min_samples_split': [14,16,20,24] #베스트 파라미터 기준 위아래 한개씩 추가후 세부조정
}

grid_cv = GridSearchCV(
     dt_clf
    ,param_grid=params
    ,scoring='accuracy'
    ,cv=5 #보통 5,홀수
    ,verbose=1
)

grid_cv.fit(X_train,y_train)
```

```python
grid_cv.best_score_
```

```python
grid_cv.best_params_
```

```python
#best model
best_grid_clf = grid_cv.best_estimator_
pred = best_grid_clf.predict(X_test)
accuracy_score(y_test,pred)
```
